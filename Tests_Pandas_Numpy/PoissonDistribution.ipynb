```python
from scipy.stats import poisson
import matplotlib.pyplot as plt
```

```python
#randon variable representing the no of resturants
#mean no of occurences of the resturant in 10 miles is 2


X=[0,1,2,3,4,5]
lmbda=2
```

```python
#probabily values 
poisson_pd = poisson.pmf(X,lmbda)
```

```python
#plot the probability distribution 

fig,ax= plt.subplots(1,1,figsize=(8,6))
ax.plot(X,poisson_pd,'bo',ms=8,label='poisson pmf')
plt.ylabel("Probability",fontsize="18")
plt.xlabel("X=No of resturants",fontsize="18")
plt.title("poisson Distribution - no of resturants vs probbility", fontsize="18")
ax.vlines(X,0,poisson_pd,colors='b',lw=5,alpha=0.5)
```
