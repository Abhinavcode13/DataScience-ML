```python
#Independent sample t test
```

```python
import random
a= [random.gauss(50,20) for x in range(30)]
b= [random.gauss(55,15) for x in range(30)]
```

```python
import seaborn as sns
sns.set_style('darkgrid')
sns.kdeplot(a,shade=True)
sns.kdeplot(b,shade=True)
```

```python
import scipy.stats as stats
```

```python
t_stats,p_value=stats.ttest_ind(a,b,equal_var=False)
```

```python
t_stats,p_value
```

```python
import numpy as np
```

```python
np.mean(a), np.mean(b)
```

```python
#Paired sample t test
```

```python
import random
a= [random.gauss(50,20) for x in range(30)]
b= [random.gauss(60,25) for x in range(30)]
```

```python
t_stat,p_value = stats.ttest_rel(a,b)
```

```python
t_stat,p_value
```

```python
np.mean(b)-np.mean(a)
```

```python
np.mean(a)
```

```python
t_stat,p_value =stats.ttest_1samp(a,100,axis=0)
```

```python
t_stat,p_value
```

```python
#z-test one sample
```

```python
import random
random.seed(20)
from statsmodels.stats.weightstats import ztest as ztest
```

```python
a =[random.gauss(100,15) for x in range (40)]
```

```python
ztest(a,value=100)
```

```python
a =[random.gauss(120,15) for x in range (40)]
```

```python
ztest(a,value=100,alternative='larger')
```

```python
 a=[random.gauss(80,15) for x in range (40)]
```

```python
ztest(a,value=100,alternative='smaller')
```

```python
#z-test two sample
```

```python
ztest(a,b,value=0)
```
